```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image,ImageOps,ImageFilter
```

```python
img = Image.open('img_512.png')
```

```python
img.size
```

```python
plt.imshow(img)
```

### GRAY SCALE

```python
gray_scale = img.convert('L')
```

```python
plt.imshow(gray_scale)
```

### Rotate

```python
img_rotated = img.rotate(90)
```

```python
plt.imshow(img_rotated)
```

### Resize

```python
img_resize = img.resize((img.width//2,img.height//2))
```

```python
plt.imshow(img_resize)
```

### Flip

```python
img_flip = img.transpose(Image.FLIP_LEFT_RIGHT)
```

```python
plt.imshow(img_flip)
```

### Black & White

```python
img_neg = img.convert('I')
```

```python
plt.imshow(img_neg)
```

### BLUR

```python
blur = img.filter(ImageFilter.BoxBlur(10))
```

```python
plt.imshow(blur)
```
