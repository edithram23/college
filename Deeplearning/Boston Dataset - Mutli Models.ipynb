```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Flatten,Dropout
from tensorflow.keras.callbacks import EarlyStopping
import numpy as np
```

```python
(x_train,y_train),(x_test,y_test) = tf.keras.datasets.boston_housing.load_data()
```

```python
(x_train,y_train),(x_test,y_test) 
```

```python
mean = x_train.mean()
x_train = x_train - mean
std = x_train.std()
x_train/=std
```

```python
x_test-= mean
x_test/= std
```

### 32 - 16 Relu Layer with 0.1LR

```python
model1 = Sequential()
model1.add(Dense(32, activation='relu', input_shape=(x_train.shape[1],)))
model1.add(Dense(16, activation='relu'))
model1.add(Dense(1))

opt = tf.keras.optimizers.SGD(learning_rate=0.1)

model1.compile(optimizer=opt,loss='MAPE')

his1 = model1.fit(x=x_train,y=y_train,epochs=1000,validation_data=(x_test,y_test))
```

```python
loss_1 = his1.history['loss'][-1]
loss_1
```

 ### 32-16 Relu with 0.01LR

```python
model2 = Sequential()
model2.add(Dense(32, activation='relu', input_shape=(x_train.shape[1],)))
model2.add(Dense(16, activation='relu'))
model2.add(Dense(1))

opt = tf.keras.optimizers.SGD(learning_rate=0.01)

model2.compile(optimizer=opt,loss='MAPE')

his2 = model2.fit(x=x_train,y=y_train,epochs=1000,validation_data=(x_test,y_test),batch_size=64)
```

```python
loss_2 = his2.history['loss'][-1]
loss_2
```

### 32Sigmoid - 16 Linear with 1e-2 LR  

```python
model3 = Sequential()
model3.add(Dense(32, activation='sigmoid', input_shape=(x_train.shape[1],)))
model3.add(Dense(16, activation='linear'))
model3.add(Dense(1))

opt = tf.keras.optimizers.SGD(learning_rate=1e-2)

model3.compile(optimizer=opt,loss='MAPE')
his3 = model3.fit(x=x_train,y=y_train,epochs=1000,validation_data=(x_test,y_test))
```

```python
loss_3 = his3.history['loss'][-1]
loss_3
```

### 32 Relu with 1e-4LR

```python
model4 = Sequential()
model4.add(Dense(32, activation='relu', input_shape=(x_train.shape[1],)))
model4.add(Dense(1))

opt = tf.keras.optimizers.SGD(learning_rate=1e-3)

model4.compile(optimizer=opt,loss='MAPE')
his4 = model4.fit(x=x_train,y=y_train,epochs=1000,validation_data=(x_test,y_test))
```

### Model 4 with Loss 20.39

#### On use of Dropout with 0.1 as the rate there is a increase in the Loss, this is inorder to reduce overfitting of the model.

```python
model_drop = Sequential()
model_drop.add(Dense(32, activation='relu', input_shape=(x_train.shape[1],)))
model_drop.add(Dropout(0.1))
model_drop.add(Dense(1))

opt = tf.keras.optimizers.SGD(learning_rate=1e-3)

model_drop.compile(optimizer=opt,loss='MAPE')
his_drop = model_drop.fit(x=x_train,y=y_train,epochs=1000,validation_data=(x_test,y_test))
```

```python
drop_loss = his_drop.history['loss'][-1]
drop_loss
```

```python
model_early = Sequential()
model_early.add(Dense(32, activation='relu', input_shape=(x_train.shape[1],)))
model_early.add(Dense(1))

es = EarlyStopping(monitor='loss',mode='min',patience=25)
opt = tf.keras.optimizers.SGD(learning_rate=1e-3)

model_early.compile(optimizer=opt,loss='MAPE')
his_early = model_early.fit(x=x_train,y=y_train,epochs=10000,validation_data=(x_test,y_test),callbacks=[es])
```

```python
early_loss = his_early.history['loss'][-1]
model_early._name = 'EarlyLoss'
early_loss
```

```python
model_early = Sequential()
model_early.add(Dense(32, activation='relu', kernel_regularizer="l1",input_shape=(x_train.shape[1],)))
model_early.add(Dense(1 ,kernel_regularizer="l2"))

es = EarlyStopping(monitor='loss',mode='min',patience=25)
opt = tf.keras.optimizers.SGD(learning_rate=1e-3)

model_early.compile(optimizer=opt,loss='MAPE')
his_early = model_early.fit(x=x_train,y=y_train,epochs=1000,validation_data=(x_test,y_test),callbacks=[es])
```

```python
early_loss = his_early.history['loss'][-1]
model_early._name = 'EarlyLoss'
early_loss
```

```python
model = [model4,model_drop,model_early]
for i in model:
    print(i.summary())
    print("Loss : ",i.history.history['loss'][-1])
```
